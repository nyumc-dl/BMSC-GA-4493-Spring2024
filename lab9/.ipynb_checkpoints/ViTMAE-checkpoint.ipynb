## Setup and Import

```python
from transformers import ViTMAEForPreTraining,  ViTMAEConfig, ViTForImageClassification,  ViTConfig
import torch
import torch.nn as nn
import torchvision
from torchvision.transforms import ToTensor, Lambda
from torchvision import datasets, transforms
from torchvision.utils import make_grid
from torchvision import transforms
from PIL import Image
import torch.optim as optim
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets, transforms
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
device
```

```python
# Define transformations
transform = transforms.Compose([
    transforms.Resize(224),  # Resize to fit the input size expected by the ViT model.
    transforms.Grayscale(num_output_channels=3),  # Convert grayscale to 3-channel.
    transforms.ToTensor(),
    # Normalize with ImageNet's mean and std
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Prepare the MNIST training dataset
train_dataset = datasets.MNIST(
    root='./data', 
    train=True, 
    download=True, 
    transform=transform
)

# Prepare the MNIST test dataset
test_dataset = datasets.MNIST(
    root='./data', 
    train=False, 
    download=True, 
    transform=transform
)

# DataLoader for the training set
train_loader = DataLoader(
    train_dataset, 
    batch_size=128, 
    shuffle=True
)

# DataLoader for the test set
test_loader = DataLoader(
    test_dataset, 
    batch_size=64, 
    shuffle=False
)
```

```python
def plot_digits(digits):
    grids = make_grid(digits, nrow=8)
    plt.figure(figsize=(12, 9))
    plt.imshow(grids.numpy()[0], cmap='gray')
```

```python
# The next() function returns the next item from the iterator.
data, label = next(iter(train_loader))
plot_digits(data)
```

## 1. Pretrain

```python
config = ViTMAEConfig()
# Load the ViT-MAE model
model = ViTMAEForPreTraining(config)
model = model.to(device)

# Define the optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=3e-4)
```

### SSL Training
Validation set is not involved in the training. However, it is beneficial to check the validation's performance even in self-supervised learning.

```python
model.train()
for epoch in range(15):  
    for i, (images, _) in enumerate(train_loader):
        images = images.to(device)
        
        outputs = model(pixel_values=images)
        loss = outputs.loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
        if i % 100 == 0:
            print(f"[Epoch {epoch+1}, Batch {i+1}] Loss: {loss.item()}")
```

```python
data, label = next(iter(test_loader))
plot_digits(data)
plt.title("Sampled original data")
```

```python
model.eval()
reconstruction = model(data.to(device)).logits
```

```python
def visualize(pixel_values, model):
    # forward pass
    outputs = model(pixel_values)
    y = model.unpatchify(outputs.logits) # (N, Channels, Height, Width)
    y = torch.einsum('nchw->nhwc', y).detach().cpu() # (N, Height, Width, Channels)
    
   

    # visualize the mask
    mask = outputs.mask.detach() #  torch.Size([64, 196,1]) 196 = 14 * 14s
    mask = mask.unsqueeze(-1).repeat(1, 1, model.config.patch_size**2 *3)  # torch.Size([64, 196, 768])
    mask = model.unpatchify(mask)  # torch.Size([64, 3, 224, 224])
    mask = torch.einsum('nchw->nhwc', mask).detach().cpu() # torch.Size([64, 224, 224, 3])
    
    x = torch.einsum('nchw->nhwc', pixel_values).detach().cpu()

    # masked image
    im_masked = x * (1 - mask)

    # MAE reconstruction pasted with visible patches
    im_paste = x * (1 - mask) + y * mask

    # make the plt figure larger
    plt.rcParams['figure.figsize'] = [24, 24]

    plt.subplot(1, 4, 1)
    show_image(x[1], "original")

    plt.subplot(1, 4, 2)
    show_image(im_masked[1], "masked")

    plt.subplot(1, 4, 3)
    show_image(y[1], "reconstruction")

    plt.subplot(1, 4, 4)
    show_image(im_paste[1], "reconstruction + visible")

    plt.show()
```

```python
def show_image(image, title=''):
    # image is [H, W, 3] 
    assert image.shape[2] == 3
    plt.imshow(torch.clip((image * imagenet_std + imagenet_mean) * 255, 0, 255).int())
    plt.title(title, fontsize=16)
    plt.axis('off')
    return
```

```python
imagenet_mean = np.array([0.485, 0.456, 0.406])
imagenet_std = np.array([0.229, 0.224, 0.225])
```

```python
device = torch.device('cpu')
model = model.to(device)
data = data.to(device)
visualize(data, model)
```

## Fine-tuning

```python
config = ViTConfig(
    image_size=224,
    num_labels=10,  
)
vit_cls = ViTForImageClassification.from_pretrained("./Vit_mae_pretrained.pth", config=config)

```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
vit_cls = vit_cls.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(vit_cls.parameters(), lr=5e-5)
```

```python
vit_cls.train()
num_epochs = 5  # Example number of epochs
print_every = 100  # Print the loss every 10 batches

for epoch in range(num_epochs):
    running_loss = 0.0
    correct_predictions = 0
    total_predictions = 0

    for i, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        
        outputs = vit_cls(images).logits
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()
        
        # Add the loss of each batch to the running loss
        running_loss += loss.item()

        # Calculate accuracy
        _, predicted = torch.max(outputs, 1)
        total_predictions += labels.size(0)
        correct_predictions += (predicted == labels).sum().item()

        # Print log
        if (i + 1) % print_every == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}], '
                  f'Step [{i+1}/{len(train_loader)}], '
                  f'Loss: {running_loss / print_every:.4f}, '
                  f'Accuracy: {100 * correct_predictions / total_predictions:.2f}%')
            
            # Reset running loss and accuracy for the next set of batches
            running_loss = 0.0
            correct_predictions = 0
            total_predictions = 0
        
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vit_cls = vit_cls.to(device)
vit_cls.eval()

# Disable gradient computation for testing (saves memory and computations)
with torch.no_grad():
    correct_predictions = 0
    total_predictions = 0

    for images, labels in test_loader:
        # Move inputs and labels to the correct device
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        outputs = vit_cls(images).logits

        # Get predictions from the maximum value
        _, predicted = torch.max(outputs, 1)
        
        # Update total and correct predictions
        total_predictions += labels.size(0)
        correct_predictions += (predicted == labels).sum().item()

    # Calculate accuracy
    accuracy = 100 * correct_predictions / total_predictions
    print(f'Accuracy on the test set: {accuracy:.2f}%')
```
